```python
import torch
import torch.nn as nn
from torch.nn import init
import torch.optim as optim
import torch.utils.data as Data
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
# 首先得到input的值 （x-values）

x = [i for i in range(1, 201, 2)]
x = np.array(x)
print(x)

#生成一个从-10到10的噪声
np.random.seed(666)   #添加一个随机种子进行伪随机，方便我之后的每次重新运行
noise = [np.random.uniform(-10, 10) for i in range(100)]


#生成y
true_w = 0.4
true_b = 3
y = true_w * x + true_b + noise

x = [[i] for i in range(1, 201, 2)]
y = [[float('{:.4f}'.format(i))] for i in y]
print(x)
print(y)

```

```python
#进行一个打乱
random.seed(666)

c = list(zip(x, y))
print(c)
random.shuffle(c)
print(c)
features, labels = zip(*c)

print(features)
print(labels)
```

```python
n_train = 70
train_features = torch.tensor(features[:n_train], dtype=torch.float)
test_features = torch.tensor(features[n_train:], dtype=torch.float)

train_labels = torch.tensor(labels[:n_train], dtype=torch.float)
test_labels = torch.tensor(labels[n_train:], dtype=torch.float)

print(train_features)
print(test_features)
print(train_labels)
print(test_labels)
```
